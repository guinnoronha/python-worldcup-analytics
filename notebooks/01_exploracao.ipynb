# 01.Bibliotecas

```python
import pandas as pd
```

# 02. Explorando tabelas

## 02.01. Tabela WorldCupMatches.csv

```python
df_matches = pd.read_csv('../data/raw/WorldCupMatches.csv')

df_matches.head()
df_matches.info()
df_matches.describe()
df_matches.isnull().sum()
```

## 02.02. Tabela WorldCups.csv

```python
df_cup = pd.read_csv('../data/raw/WorldCups.csv')

df_cup.head()
df_cup.info()
df_cup.describe()
df_cup.isnull().sum()
```

## 02.03. Join entre tabelas

```python
df_full = df_matches.merge(
    df_cup,
    how='left',
    on='Year',
    suffixes=('_matches', '_cup')
)

df_full.head()
df_full.info()
df_full.describe()
df_full.isnull().sum()
```

# 03. Export base completa para .csv

```python
df_full.to_csv('../data/full/WorldCupFull.csv', index=False)
print('Arquivo WorldCupFull.csv gerado com sucesso!')
```

# 04. Descritivo de Tipo de Colunas para a Base Completa

```python
tipos_colunas = pd.DataFrame({
    'coluna': df_full.columns,
    'tipo': df_full.dtypes.values
})

tipos_colunas
```
